Definimos una función para extrar las etiquetas, el index de las mismas y la señal de cualquier registro de ECG

````python
def load_ecg(file):    
    # Obtener SEÑAL ECG
    record = wfdb.rdrecord(file)
    # Obtener anotaciones
    annotation = wfdb.rdann(file, 'atr')
    
    # Extraemos la señal
    p_signal = record.p_signal

    # Se extra el simbolo y el numero de muestra (index)
    atr_sym = annotation.symbol
    atr_sample = annotation.sample
    
    return p_signal, atr_sym, atr_sample
````

Definimos una función para elaborar el dataset bajo las primicia de obtener las etiquetas y una cantidad de muestras n_segundos antes y n_segundos después, generando datos como el siguiente donde:

* Se alamcena la etiqueta objetivo (central)
* Se almacena el index de la etiqueta objetivo

* Se almacena la información de mV de la señal como parametro para X
*Utiliza como parametros de entrada, la cantidad de segundos, la tabla de valores anormales, la frecuencia de registro (previamente mencionada) en la descripción, los nombres de los archivos o "pacientes"


En esta función solo se extraen los datos, posteriormente se conjuntan en una sola variable.

````python

def make_dataset(pts, num_sec, fs, abnormal):

    # Inicializamos arreglos
    num_cols = 2*num_sec * fs #Numero de columnas = los segundos antes y después que deseamos
    # recuperar por la frecuencia 
    X_all = np.zeros((1,num_cols)) #Arreglo para almacenar valores de datos en X  
    Y_all = np.zeros((1,1)) #Arreglo para almacenar valores de datos en Y
    sym_all = [] #Lista de todos los simbolos
    
    # Lista para llevar registro de todos los latidos a través de los registros de pacientes
    max_rows = []
    
    for archive in rdnames:
        file = dl_dir + '/' + archive
        p_signal, atr_sym, atr_sample = load_ecg(file)
        
        # Recuperamos solo la señal de la II derivada
        p_signal = p_signal[:,0]
        
        # Excluimos la categoria de no latidos, en base a los grupos previamente formados
        df_ann = pd.DataFrame({'atr_sym':atr_sym,
                              'atr_sample':atr_sample})
        df_ann = df_ann.loc[df_ann.atr_sym.isin(abnormal + ['N'])]
        
        X,Y,sym = build_XY(p_signal,df_ann, num_cols, abnormal) 
        sym_all = sym_all+sym
        max_rows.append(X.shape[0]) 
        X_all = np.append(X_all,X,axis = 0) #Se añaden elementos a la lista
        Y_all = np.append(Y_all,Y,axis = 0) #Sea añaden elementos a la lista
        
    # Se almacena, los datos obtenidos exceptuando la primer iteraccion donde se inicializo la lista
    X_all = X_all[1:,:]
    Y_all = Y_all[1:,:]

    return X_all, Y_all, sym_all
````

---

Esta función conjunta los datos extraidos de la funcion make_dataset(). 

Como parametros de entrada utiliza los datos recolectados por make_dataset() y genera las etiquetas para introducir a la red

Como ejemplo de visualización se generan imagenes como la de abajo por cada numero de etiquetas presente

image.png

````python
def build_XY(p_signal, df_ann, num_cols, abnormal):
    # Elabora la matriz para el entrenamiento del dataset
    #se determina el numero de filas segun la longitud del dataset de anotaciones
    num_rows = len(df_ann)
    #Los datos X, se inicializan
    X = np.zeros((num_rows, num_cols))
    #Los datos en Y se inicializan
    Y = np.zeros((num_rows,1))
    sym = []
    
    # Se mantiene un conteo de las filas
    max_row = 0
    #Se genera una muestra de tiempo por cada etiqueta presente
    for atr_sample, atr_sym in zip(df_ann.atr_sample.values,df_ann.atr_sym.values):

        left = max([0,(atr_sample - num_sec*fs) ]) #Se genera la muestra en n_segundos antes de la anotación
        right = min([len(p_signal),(atr_sample + num_sec*fs) ]) #Se genera la muestra en n_segundos después de la anotación
        x = p_signal[left: right] #Se genera una lista temporal, para almacenar los valores de X
        if len(x) == num_cols:
            X[max_row,:] = x #Se generan datos X
            Y[max_row,:] = int(atr_sym in abnormal) #Se genera datos Y
            sym.append(atr_sym) #Se añade simbolo a la lista
            max_row += 1 #Se aumenta contador
    X = X[:max_row,:]
    Y = Y[:max_row,:]
    return X,Y,sym
````

Metricas

````python
#Realizamos reporte de estadisticas
def print_report(y_actual, y_pred, thresh):
    # Function to print evaluation metrics
    auc = roc_auc_score(y_actual, y_pred)
    accuracy = accuracy_score(y_actual, (y_pred > thresh))
    recall = recall_score(y_actual, (y_pred > thresh))
    precision = precision_score(y_actual, (y_pred > thresh))
    specificity = sum((y_pred < thresh) & (y_actual == 0)) /sum(y_actual ==0)
    prevalence = (sum(y_actual)/len(y_actual))
    print('AUC:%.3f'%auc)
    print('Accuracy:%.3f'%accuracy)
    print('Recall:%.3f'%recall)
    print('Precision:%.3f'%precision)
    print('Specificity:%.3f'%specificity)
    print('Prevalence:%.3f'%prevalence)
    print(' ')
    return auc, accuracy, recall, precision, specificity
````

````python
#Definimos el umbral...
thresh = (sum(y_train)/len(y_train))[0]
````

````python
# Metricas
print('Train');
print_report(y_train, y_train_preds_cnn, thresh)
print('Valid');
print_report(y_test, y_test_preds_cnn, thresh);
````
